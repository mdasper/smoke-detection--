```python
import gradio as gr
from ultralyticsplus import YOLO, render_result

# Load model
model = YOLO('kittendev/YOLOv8m-smoke-detection')

# Set model parameters
model.overrides['conf'] = 0.25  # NMS confidence threshold
model.overrides['iou'] = 0.45  # NMS IoU threshold
model.overrides['agnostic_nms'] = False  # NMS class-agnostic
model.overrides['max_det'] = 1000  # maximum number of detections per image

# Define function to process the uploaded image
def detect_smoke(image):
    # Perform inference
    results = model.predict(image)

    # Get results and render
    render = render_result(model=model, image=image, result=results[0])
    return render

# Create Gradio interface
iface = gr.Interface(
    fn=detect_smoke,  # Function that runs on image input
    inputs=gr.Image(type="pil"),  # Accepts image upload
    outputs=gr.Image(type="pil"),  # Outputs rendered image
    live=True  # Enable live prediction
)

# Launch the Gradio interface
iface.launch()
```
